## Завдання 2. Аналіз повноти відповідей
Визнач, скільки респондентів відповіли на всі запитання опитування. Для цього тобі знадобиться:

Отримати список питань з файлу survey_results_schema (поле "qname")
Перевірити наявність відповідей на ці питання у кожного респондента
Підрахувати кількість респондентів без пропущених значень
Примітка! Список питань міститься у файлі survey_results_schema, то ж треба перевірити, чи була відповідь у опитуванні на запитання, яке знаходиться у survey_results_schema, поле "qname". Для фільтрації лише цих значень можеш використати метод інтерсекції двох множин.

```python
# 1. Створюємо множину унікальних назв питань із файлу схеми (поле qname)
questions_set = set(schema['qname'].unique())

# 2. Отримуємо назви стовбців основного датафрейму
df_columns_set = set(df.columns)

# 3. Знаходимо перетин множин (intersection), щоб залишити тільки ті питання, які реально існують у стовбцях df
common_questions = list(questions_set.intersection(df_columns_set))

# 4. Використовуємо метод .dropna() для фільтрації рядків, де є хоча б один пропуск у вибраних колонках
full_responses_df = df[common_questions].dropna()

# 5. Використовуємо .shape[0], щоб отримати кількість таких респондентів
count_full_responses = full_responses_df.shape[0]

print(f"Кількість респондентів, які відповіли на всі запитання: {count_full_responses}")
```

## Завдання 3. Статистичний аналіз досвіду респондентів
Обчисли міри центральної тенденції для поля WorkExp (досвід роботи респондентів). Тобі потрібно знайти:
* Середнє значення (mean)
* Медіану (median)
* Моду (mode)

```python
# Обчислюємо міри центральної тенденції
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0] # mode повертає Series, беремо перший елемент

stats_exp = pd.Series({
    'Mean': mean_exp,
    'Median': median_exp,
    'Mode': mode_exp
})

print("Статистика досвіду роботи (WorkExp):")
print(stats_exp)
```

# Завдання 5. Визначення популярності Python
Обчисли відсоток респондентів, які програмують на Python. Зверни увагу, що поле з мовами програмування може містити множинні значення, тому потрібно перевірити наявність Python в цьому полі.

```python
# 1. Створюємо новий стовпець 'worked_with_python'
# Використовуємо .str.contains з параметрами:
# 'Python' — що шукаємо
# na=False — замінюємо пропуски (NaN) на False
# case=False — ігноруємо регістр (щоб знайшло і python, і Python)
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False, case=False)

# 2. Рахуємо кількість респондентів, які працювали з Python
# Оскільки стовпець містить True/False, .sum() порахує тільки True як одиниці
python_users_count = df['worked_with_python'].sum()

# 3. Обчислюємо частку у відсотках
# Ділимо кількість Python-користувачів на загальну кількість рядків і множимо на 100
python_percentage = (python_users_count / df.shape[0]) * 100

# 4. Виводимо результат
print(f"Відсоток респондентів: {python_percentage:.1f}%")
```

## Завдання 6. Аналіз шляхів навчання програмуванню
Визнач, скільки респондентів навчалося програмувати за допомогою онлайн курсів. Знайди відповідне поле в датасеті, яке описує способи навчання, та підрахуй тих, хто обрав онлайн курси.


```python

# 1. Створюємо новий стовпець 'Learned_with_online_courses'
# Шукаємо значення 'Online Courses' у стовпці 'LearnCode'
# Параметр na=False гарантує, що порожні значення стануть False
# case=False робить пошук нечутливим до регістру
df['Learned_with_online_courses'] = df['LearnCode'].str.contains('Online Courses', na=False, case=False)

# 2. Підраховуємо кількість респондентів, які обрали онлайн-курси
# Метод .sum() додасть усі True (як 1) та ігноруватиме False (як 0)
online_courses_count = df['Learned_with_online_courses'].sum()

# 3. Виводимо фінальний результат
print(f"Кількість респондентів, які навчалися через онлайн курси: {online_courses_count}")
```

## Завдання 7. Географічний аналіз компенсації Python-розробників
Серед респондентів, що програмують на Python, проведи аналіз компенсації (ConvertedCompYearly) в розрізі країн. Для кожної країни обчисли:
* Середню суму компенсації
* Медіанну суму компенсації




```python
# 1. Відбираємо лише тих, хто працював з Python (використовуємо стовпець із Завдання 5)
# Також відразу видаляємо рядки, де відсутнє значення зарплати (ConvertedCompYearly)
python_devs_with_salary = df[df['worked_with_python'] & df['ConvertedCompYearly'].notna()].copy()

# 2. Групкуємо дані за країною (Country)
# та обчислюємо середню (mean) і медіанну (median) зарплату
py_progers_by_country = python_devs_with_salary.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).round(2)

# 3. Виводимо таблицю з результатами
# Додамо сортування за медіаною, щоб побачити найбільш оплачувані країни
display(py_progers_by_country.sort_values(by='median', ascending=False).head(10))
```

## Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів
Визнач рівні освіти 5 респондентів з найбільшою компенсацією. Тобі потрібно:
* Відсортувати респондентів за рівнем компенсації
* Вибрати топ-5
* Показати їх рівні освіти

```python
# 1. Створюємо вибірку лише з трьох необхідних стовпців
top_earners_data = df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]

# 2. Сортуємо дані за колонкою ConvertedCompYearly у порядку спадання (від більшого до меншого)
# 3. Оновлюємо індекси за допомогою .reset_index(drop=True), щоб вони йшли від 0 до 4
# 4. Обираємо перші 5 рядків за допомогою .head(5)
top_5_specialists = top_earners_data.sort_values(by='ConvertedCompYearly', ascending=False).reset_index(drop=True).head(5)

# 5. Виводимо результат
print("Топ-5 респондентів з найбільшою зарплатою та їхня освіта:")
display(top_5_specialists)
```

## *Завдання 9. Аналіз популярності Python по віковим категоріям
Для кожної вікової категорії визнач відсоток респондентів, які програмують на Python. Це дасть уявлення про те, в яких вікових групах Python є найпопулярнішим.

```python
# Групуємо за віком, для кожної групи рахуємо відсоток тих, у кого є Python
def get_python_pct(group):
    return (group.str.contains('Python', na=False).sum() / len(group)).round(2) * 100

age_python_popularity = df.groupby('Age')['LanguageHaveWorkedWith'].apply(get_python_pct).reset_index()
age_python_popularity.columns = ['Вікова категорія', 'Відсоток Python (%)']

display(age_python_popularity)
```

## *Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників


```python
# 1. Знаходимо поріг 75-го перцентиля
q75 = df['ConvertedCompYearly'].quantile(0.75)

# 2. Фільтруємо: компенсація > q75 ТА RemoteWork == 'Remote'
high_earners_remote = df[(df['ConvertedCompYearly'] > q75) & (df['RemoteWork'] == 'Remote')]

# 3. Рахуємо найпопулярніші індустрії (колонка Industry)
# Оскільки може бути кілька значень, використовуємо value_counts
top_industries = high_earners_remote['Industry'].value_counts()

print(f"Найпопулярніші індустрії серед високооплачуваних віддалених працівників (Поріг: ${q75:,.0f}):")
print(top_industries.head(10))
```
